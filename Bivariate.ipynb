# Replacing the null values using imputer

```python
# Reading the dataset 

data = pd.read_csv("Placement.csv")
data
```

```python
# Checking for null values

data.isna().sum()
```

```python
# Splitting the qualitative and quantitative columns using the function we created

from univariate import univariate
uni = univariate()
Qual,Quan = uni.Qualquan(data)
```

```python
Qual
```

```python
Quan
```

```python
# Filling the missing values using the imputer

from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = 'constant')
imp.fit(data[Quan])
df = imp.transform(data[Quan])
df = pd.DataFrame(df, columns = Quan)
df
```

```python
# Checking for update

df.isna().sum()
```

```python
df1 = data[Qual]
df1
```

```python
frame = [df,df1]
```

```python
processed_data = pd.concat(frame, axis = 1)
processed_data
```

```python
processed_data.to_csv("Preporcessed_Plcement.csv", index = False)
```

# Finding the covariance and correlation between columns

```python
cleandata.cov()
```

```python
plt.scatter(cleandata['degree_p'], cleandata['etest_p'], color = 'b')
slope, intercept = np.polyfit(cleandata['degree_p'], cleandata['etest_p'], 1)
plt.plot(cleandata['degree_p'], slope*np.array(cleandata['degree_p']) + intercept, color='red', label='Linear Regression Line')
plt.xlabel("degree_p")
plt.ylabel("etest_p")
plt.title("Covariance between degree_p and etest_p")
plt.show()

```

There is a positive Covariace between the degree pass mark(degree_p) and the entrance pass mark(etest_p) with value 21.929469

```python
plt.scatter(cleandata['etest_p'], cleandata['mba_p'], color = 'b')
slope, intercept = np.polyfit(cleandata['etest_p'], cleandata['mba_p'], 1)
plt.plot(cleandata['etest_p'], slope*np.array(cleandata['etest_p']) + intercept, color='red', label='Linear Regression Line')
plt.xlabel("etest_p")
plt.ylabel("mba_p")
plt.title("Covariance between mba_p and etest_p")
plt.show()
```

There is a positive covariance between the entrance pass mark and MBA pass mark with value 19.886973

```python
cleandata.corr()
```

```python
plt.scatter(cleandata['mba_p'], cleandata['salary'], color = 'b')
slope, intercept = np.polyfit(cleandata['mba_p'], cleandata['salary'], 1)
plt.plot(cleandata['mba_p'], slope*np.array(cleandata['mba_p']) + intercept, color='red', label='Linear Regression Line')
plt.xlabel("mba_p")
plt.ylabel("salary")
plt.title("Correlation beween mba_p and salary")
plt.show()
```

There is a positive correlation between the MBA pass mark and salary with value 0.139823
